# 1 创建数组

## 1.1 使用Python序列创建数组

```python
lst_a  = [1, 2, 3]
arr1 = np.array(lst_a)
print(arr1)
print(type(arr1))
```

```python
dic_a = {
    "name" : "kobayashi",
    "age" : 19,
}

arr2 = np.array(dic_a)
print(arr2)
print(type(arr2))
```

## 2 NumPy 自动生成数组

1. arange 创建顺序数组
2. zeros 创建元素全为 0 的数组
3. ones 创建元素全为 1 的数组
4. eye 创建对角数组
5. random 随机生成数组
6. linspace 生成均匀间隔数组

```python
# arange 函数
arr3 = np.arange(10)
arr3
```

```python
# zeros 函数
arr4 = np.zeros(10)
arr4
```

```python
# ones 函数
arr5 = np.ones((2, 3))
arr5
```

```python
# eye 函数
arr6 = np.eye(3)
arr6
```

```python
# random 函数

# 从 [0, 1) 中随机取值
arr7 = np.random.rand(2, 3)
print(arr7)

# 创建指定区间的随机整数数组
arr8 = np.random.randint(1, 10, (2, 3))
print(arr8)

# 创建指定区间的随机浮点数数组
arr9 = np.random.uniform(1, 10, (2, 3))
print(arr9)
```

```python
# linspace 函数

# 均匀比例的数组
arr10 = np.linspace(-1, 1, 20)
arr10
```

```python
# empty 函数
arr11 = np.empty((2, 3))
arr11
```

# 3 数组的拼接与分割

## 1 数组拼接

1. 横向拼接 hstack
2. 纵向拼接 vstack

```python
# 横向拼接
arr3 = np.arange(20).reshape(4, 5)
arr3
```

```python
np.hstack((arr2, arr3))
```

```python
np.vstack((arr2, arr3))
```

```python
# 水平分割
np.hsplit(arr3, 5)
```

```python
np.vsplit(arr2, 2)
```
